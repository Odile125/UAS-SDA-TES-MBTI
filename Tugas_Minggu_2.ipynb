```python
class CircularQueue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.queue = [None] * capacity
        self.tail = -1
        self.head = 0
        self.size = 0

    def enqueue(self, item):
        if self.size == self.capacity:
            print("Error: Queue is Full")
        else:
            self.tail = (self.tail + 1) % self.capacity
            self.queue[self.tail] = item
            self.size += 1

    def dequeue(self):
        if self.size == 0:
            print("Error: Queue is Empty")
            return None
        else:
            item = self.queue[self.head]
            self.head = (self.head + 1) % self.capacity
            self.size -= 1
            return item

    def display(self):
        if self.size == 0:
            print("Queue is Empty\n")
        else:
            index = self.head
            for i in range(self.size):
                print(self.queue[index])
                index = (index + 1) % self.capacity
```

```python
# Pengujian CircularQueue dengan kapasitas 4
cq = CircularQueue(4)
cq.enqueue(4)
cq.enqueue(3)
cq.enqueue(2)
cq.enqueue(1)

print("Isi antrian:")
cq.display()
# Hapus elemen pertama
print("Dequeue 1 elemen:", cq.dequeue())
# Hapus elemen 2
print("Dequeue 1 elemen:", cq.dequeue())
# Tambahkan elemen pertama
print("Tambahkan elemen D:")
cq.enqueue("D")
# Tambahkan elemen kedua
print("Tambahkan elemen E:")
cq.enqueue("E")

print("Isi antrian sekarang:")
cq.display()

```

```python
from datetime import datetime

class PriorityQueueUltah:
    def __init__(self):
        self.queue = []

    def days_until_birthday(self, month, day):
        today = datetime.now()
        this_year = today.year
        birthday = datetime(this_year, month, day)

        if birthday < today:
            birthday = datetime(this_year + 1, month, day)

        return (birthday - today).days

    def insert(self, name, month, day):
        days = self.days_until_birthday(month, day)
        self.queue.append((name, days))
        print(f"Ditambahkan: {name} (Ulang tahun dalam {days} hari)")

    def delete(self):
        if not self.queue:
            print("Queue kosong!")
            return None

        # Cari yang paling dekat ulang tahunnya
        min_index = 0
        for i in range(1, len(self.queue)):
            if self.queue[i][1] < self.queue[min_index][1]:
                min_index = i

        removed = self.queue.pop(min_index)
        print(f"Dihapus: {removed[0]} (Ulang tahun dalam {removed[1]} hari)")
        return removed

    def display(self):
        if not self.queue:
            print("Queue kosong!")
        else:
            print("Daftar Antrian (berdasarkan hari menuju ulang tahun):")
            for name, days in self.queue:
                print(f"- {name} (dalam {days} hari)")


```

```python
# Tes
pq = PriorityQueueUltah()
pq.insert("By", 2, 1)
pq.insert("Dinda", 4, 20)
pq.insert("Kevin", 5, 5)
pq.insert("Naya", 12, 25)
pq.insert("Rafi", 10, 10)
pq.display()

print("\n--- Delete Dua Orang dengan ulang tahun terdekat ---")
pq.delete()
pq.delete()
pq.display()
```
