```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Contoh Data
# Misalkan kita memiliki dataset hasil tes MBTI yang sudah diproses.
# Data ini hanya sebagai contoh:
data = {
    'Ekstrovert': [1, 0, 1, 0, 1, 0, 1, 0],
    'Intuisi': [1, 1, 0, 0, 1, 0, 1, 0],
    'Berpikir': [1, 0, 1, 0, 0, 1, 1, 0],
    'Penilaian': [1, 1, 0, 0, 1, 0, 1, 0],
    'Tipe': ['ENTJ', 'INFJ', 'ESTJ', 'ISFJ', 'ENTJ', 'INFJ', 'ESTJ', 'ISFJ']
}

# Buat DataFrame
df = pd.DataFrame(data)

# Misalkan feature (X) adalah hasil tes psikologi dan label (y) adalah tipe kepribadian
X = df[['Ekstrovert', 'Intuisi', 'Berpikir', 'Penilaian']]
y = df['Tipe']

# Pisahkan data untuk training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inisialisasi model decision tree classifier
clf = DecisionTreeClassifier(criterion='entropy', random_state=42)

# Latih model dengan data training
clf.fit(X_train, y_train)

# Lakukan prediksi pada data testing
y_pred = clf.predict(X_test)

# Evaluasi model
print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Visualisasi sederhana decision tree dalam format teks
tree_rules = export_text(clf, feature_names=list(X.columns))
print("\nDecision Tree Rules:\n", tree_rules)
```

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# Contoh dataset sederhana: setiap baris adalah jawaban user terhadap 4 dimensi MBTI
# 1: YA, 0: TIDAK
# Fitur: [Introvert?, Sensing?, Thinking?, Judging?]
data = [
    [1, 1, 1, 1],  # ISTJ
    [1, 1, 0, 1],  # ISFJ
    [1, 0, 1, 1],  # INTJ
    [1, 0, 0, 1],  # INFJ
    [0, 1, 1, 1],  # ESTJ
    [0, 1, 0, 1],  # ESFJ
    [0, 0, 1, 1],  # ENTJ
    [0, 0, 0, 1],  # ENFJ
    [0, 0, 0, 0],  # ENFP
    [1, 0, 0, 0],  # INFP
]

# Label yang sesuai (target)
labels = [
    "ISTJ", "ISFJ", "INTJ", "INFJ", 
    "ESTJ", "ESFJ", "ENTJ", "ENFJ", 
    "ENFP", "INFP"
]

# Buat DataFrame
df = pd.DataFrame(data, columns=["Introvert", "Sensing", "Thinking", "Judging"])

# Latih model Decision Tree
clf = DecisionTreeClassifier()
clf.fit(df, labels)

# Tes input dari user (misal: [Introvert=1, Sensing=0, Thinking=1, Judging=1])
user_input = [[1, 0, 1, 1]]
prediction = clf.predict(user_input)

print("Hasil tipe kepribadian MBTI Anda adalah:", prediction[0])
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Data dummy untuk MBTI
X = [[1, 1, 1, 1], [1, 0, 0, 0], [0, 1, 0, 1]]
y = ['ISTJ', 'INFP', 'ESFJ']

# Latih model decision tree
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Visualisasi pohon keputusan
plt.figure(figsize=(10,6))
plot_tree(clf, feature_names=["I", "S", "T", "J"], class_names=clf.classes_, filled=True)
plt.title("Decision Tree MBTI")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import matplotlib.pyplot as plt

# Dataset MBTI sederhana
data = [
    [1, 1, 1, 1],  # ISTJ
    [1, 1, 0, 1],  # ISFJ
    [1, 0, 1, 1],  # INTJ
    [1, 0, 0, 1],  # INFJ
    [0, 1, 1, 1],  # ESTJ
    [0, 1, 0, 1],  # ESFJ
    [0, 0, 1, 0],  # ENTP
    [0, 0, 0, 0],  # ENFP
]
labels = ["ISTJ", "ISFJ", "INTJ", "INFJ", "ESTJ", "ESFJ", "ENTP", "ENFP"]

# Membuat DataFrame
df = pd.DataFrame(data, columns=["Introvert", "Sensing", "Thinking", "Judging"])

# Melatih model Decision Tree
clf = DecisionTreeClassifier(random_state=42)
clf.fit(df, labels)

# Visualisasi pohon keputusan
plt.figure(figsize=(16, 10))
plot_tree(
    clf,
    feature_names=["Introvert", "Sensing", "Thinking", "Judging"],
    class_names=clf.classes_,
    filled=True,
    rounded=True,
    precision=2
)
plt.title("Visualisasi Pohon Keputusan MBTI (Lebih Rapi)")
plt.show()
```
