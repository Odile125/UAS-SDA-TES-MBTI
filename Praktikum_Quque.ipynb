# Materi Queue

## Struktur data queue ( dibaca : kiu ) adalah salah satu struktur data linear yang memiliki prinsip  
## FIFO (First In, First Out), yang artinya data yang masuk pertama akan keluar pertama

image.png

## Konsep Singkat

enqueue(value) → menambahkan data ke belakang antrian

dequeue() → menghapus data dari depan antrian

## Implementasi Queue Pada Pyhton

````python
class Queue1:
    def __init__(self):
        self.list = []

    def __str__(self):
        if self.isEmpty():
            return "The queue is empty."
        values = [str(x) for x in self.list]
        return ' '.join(values)

    def isEmpty(self):
        return self.list == []

    def enqueue(self, value):
        self.list.append(value) # Fungsi enquque

    def dequeue(self):
        if self.isEmpty():
            return "The queue is empty."
        else:
            return self.list.pop(0) # Fungsi dequququ

    def peek(self):
        if self.isEmpty():
            return "The queue is empty."
        else:
            return self.list[0]

    def delete(self):
        self.list = None
        
    def sort_alpabeth_ass(self):
        self.list.sort()
        
    def copy(self):
        return self.list
````

## Contoh Penggunaan

## Inisiasi Queue

````python
q = Queue1()
print(q)
print("Queue kosong :", q.isEmpty())
````

## Masukkan Isi Data Queue

````python
q.enqueue("A")
q.enqueue("B")
q.enqueue("C")
print("Isi queue:", q)
````

## Hapus Isi Queue

````python
print("Dequeued:", q.dequeue())
print("Isi queue setelah dequeue:", q)
````

## Cek Elemen Terdepan Pada Queue

````python
print("Elemen di depan:", q.peek())
````

# Circular Queue


## Circular Queue memutar ke awal array untuk memanfaatkan ruang kosong—mirip seperti jalur melingkar.

image.png

````python
class CircularQueue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.queue = [None] * capacity
        self.tail = -1
        self.head = 0
        self.size = 0

    def enqueue(self, item):
        if self.size == self.capacity:
            print("Error: Queue is Full")
        else:
            self.tail = (self.tail + 1) % self.capacity
            self.queue[self.tail] = item
            self.size += 1

    def dequeue(self):
        if self.size == 0:
            print("Error: Queue is Empty")
            return None
        else:
            item = self.queue[self.head]
            self.head = (self.head + 1) % self.capacity
            self.size -= 1
            return item

    def display(self):
        if self.size == 0:
            print("Queue is Empty\n")
        else:
            index = self.head
            for i in range(self.size):
                print(self.queue[index])
                index = (index + 1) % self.capacity
````

## Contoh Penggunaan Circular Queue

## Inisiasi Circular Queue

````python
cq = CircularQueue(3) # Deklarasi awal beserta size
print("Queue Circular Kosong")
cq.display()
````

## Input Data Pada Circular Queue

````python
cq.enqueue("A")
cq.enqueue("B")
cq.enqueue("C")

print("Isi antrian:")
cq.display()
````

## Hapus Data Circular Queue

````python
print("Dequeue 1 elemen:", cq.dequeue())

print("Tambahkan elemen D:")
cq.enqueue("D")

print("Isi antrian sekarang:")
cq.display()
````

# Priority Queue

## Priority Queue adalah struktur data di mana setiap elemen memiliki prioritas. Elemen dengan prioritas tertinggi akan keluar lebih dulu, bukan berdasarkan urutan masuk (FIFO).



## Implementasi Priority Queue di Python


````python
class PriorityQueue:
    def __init__(self):
        self.queue = []

    def __str__(self):
        return ' '.join([str(i) for i in self.queue])

    def isEmpty(self):
        return len(self.queue) == 0

    def insert(self, data):
        self.queue.append(data)

    def delete(self):
        try:
            max_val = 0
            for i in range(len(self.queue)):
                if self.queue[i] > self.queue[max_val]:
                    max_val = i
            item = self.queue[max_val]
            del self.queue[max_val]
            return item
        except IndexError:
            print("Queue kosong!")
            return None
````

## Inisiasi Priority Queue

````python
pq = PriorityQueue()
print("Queue Prioritas Kosong:", pq.isEmpty())
````

## Masukkan Elemen Pada Priority Queue

````python
pq.insert(12)
pq.insert(1)
pq.insert(14)
pq.insert(7)

print("Isi priority queue:", pq)
````

## Hapus Data Priority Pada Queue

````python
print("Elemen prioritas tertinggi (dihapus):", pq.delete())
print("Elemen prioritas tertinggi (dihapus):", pq.delete())
print("Setelah delete:", pq)
````

# Queue Didalam Linked List

🧠 Penjelasan Konsep untuk Mahasiswa

enqueue:
* Tambah node ke tail.
* Jika queue kosong, maka head dan tail menunjuk ke node baru.

dequeue:
* Ambil elemen dari head.
* Geser head ke node berikutnya.
* Jika setelah dequeue, queue kosong, maka tail juga di-reset ke None.

Kenapa pakai Linked List?
Tidak terbatas ukuran.

````python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None

    def __str__(self):
        return str(self.value)


class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next


class Queue:
    def __init__(self):
        self.LinkedList = LinkedList()

    def __str__(self):
        values = [str(x.value) for x in self.LinkedList]
        return '\n'.join(values)

    def enqueue(self, value):
        new_node = Node(value)
        if self.LinkedList.head is None:
            self.LinkedList.head = new_node
            self.LinkedList.tail = new_node
        else:
            self.LinkedList.tail.next = new_node
            self.LinkedList.tail = new_node

    def dequeue(self):
        if self.LinkedList.head is None:
            print("Queue is empty")
            return None
        else:
            dequeued_value = self.LinkedList.head.value
            self.LinkedList.head = self.LinkedList.head.next
            if self.LinkedList.head is None:
                self.LinkedList.tail = None
            return dequeued_value
````

````python
q = Queue()
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

print("Isi Queue:")
print(q)

print("\nDequeued:", q.dequeue())
print("Setelah dequeue:")
print(q)
````

# Studi Kasus: Prioritas Pelayanan Pasien Berdasarkan Umur

````python
class Pasien:
    def __init__(self, nama, umur, keluhan):
        self.nama = nama
        self.umur = umur
        self.keluhan = keluhan

    def __str__(self):
        return f"{self.nama} ({self.umur} tahun) - {self.keluhan}"


class PriorityQueue:
    def __init__(self):
        self.queue = []

    def isEmpty(self):
        return len(self.queue) == 0

    def insert(self, pasien):
        self.queue.append(pasien)

    def delete(self):
        if self.isEmpty():
            print("Antrian kosong!")
            return None
        # Cari pasien dengan umur tertua
        max_index = 0
        for i in range(1, len(self.queue)):
            if self.queue[i].umur > self.queue[max_index].umur:
                max_index = i
        return self.queue.pop(max_index)

    def __str__(self):
        if self.isEmpty():
            return "Antrian kosong"
        return '\n'.join([str(p) for p in self.queue])
````

````python
antrian_prioritas = PriorityQueue()

# Menambahkan pasien
antrian_prioritas.insert(Pasien("Pak Budi", 72, "Sesak napas"))
antrian_prioritas.insert(Pasien("Ibu Sari", 65, "Pusing"))
antrian_prioritas.insert(Pasien("Andi", 99, "Demam"))
````

````python

print("📋 Daftar Antrian Pasien:")
print(antrian_prioritas)

print("\n🔔 Melayani pasien prioritas (umur tertua):")
pasien_dilayani = antrian_prioritas.delete()
print("Sedang dilayani:", pasien_dilayani)

print("\n📋 Sisa Antrian:")
print(antrian_prioritas)
````
